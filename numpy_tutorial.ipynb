# Broadcasting

How numpy treats arrays with different shapes

- N*M dim array, broadcast a single scalar value
- N*M dim array, broadcast a vector of (NX1)
- N*M dim array, broadcast a vector of (1XN)

# Vectorization

```python
arr = np.array([-10,2,5,-5,10])

def get_square(number):
    if number < 0: #Negative
        return 0
    else:
        return number ** 2
```

```python
vectorized_square = np.vectorize(get_square)
vectorized_square
```

```python
from math import pi
```

```python
#User Defines the function
def area_of_a_circle(radius):
    return pi * radius * radius

#User array
radius_arr = np.array([10,100,5])

#Vectorized user function
vectorize_get_area_of_circle = np.vectorize(area_of_a_circle)

#Vectorized user function is invoked
vectorize_get_area_of_circle(radius_arr)
```
