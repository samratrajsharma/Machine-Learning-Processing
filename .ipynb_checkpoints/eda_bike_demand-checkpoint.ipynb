# Person Correlation

Range : -1 to +1

- `-1`: Negative pearson colrelation means increase in value of 1 variable is associated with decrease in value of another.
- `0`: No correaltion
- `1`: Positive Corelation, i.e. one variable increaes then the other variable also increases

Country - Recession Index
Sales of Iphone
