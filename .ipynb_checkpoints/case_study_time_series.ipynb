```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("./rossmann-store-sales/train.csv",
                parse_dates=["Date"], # Specifically telling pandas use `Date` as a date time variable
                #index_col="Date"
                ) #Data frame index as a Date
df.head()
```

```python
store = pd.read_csv("./rossmann-store-sales/store.csv")
store.head()
```

# Extract the time components from date

```python
store1 = df[df["Store"]==1]
store1.head()
```

```python
store1["month"] = store1["Date"].dt.month
store1["month"].head()
```

## Month as a feature
January=1, December=12.

```python
store1["month"].value_counts()
```

```python
sns.boxplot(x="month", y="Sales",data=store1)
```

## Year as a feature

```python
store1["year"] = store1["Date"].dt.year
store1["year"].head()
```

```python
sns.boxplot(x="year", y="Sales",data=store1)
```

```python
sns.boxplot(x="year", y="Sales",hue="month",data=store1)
```

## Extract Day of Week

Monday=0, Sunday=6.

```python
store1["day_of_week"] = store1["Date"].dt.dayofweek
store1["day_of_week"].head()
```

```python
sns.boxplot(x="day_of_week", y="Sales",data=store1)
```

```python
help(seasonal_decompose)
```
